# Deep Learning algorithms

# 1. 1D-CNN

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.preprocessing import StandardScaler

# Assuming X_train and X_test have shapes (326, 1600) and (82, 1600), respectively

# Reshape X_train and X_test to match the assumed input shape of the model
scaler1 = StandardScaler()
X_train_scaled2 = scaler1.fit_transform(X_train_9_all.values.reshape(X_train_9_all.shape[0], -1))
X_test_scaled2 = scaler1.transform(X_test_9_all.values.reshape(X_test_9_all.shape[0], -1))

X_train_9_all_reshaped = np.reshape(X_train_scaled2, (X_train_9_all.shape[0], 1600, 1))
X_test_9_all_reshaped = np.reshape(X_test_scaled2, (X_test_9_all.shape[0], 1600, 1))

#X_train_1_reshaped = np.reshape(X_train_scaled1.values, (X_train_1.shape[0], 1600, 1))
#X_test_1_reshaped = np.reshape(X_test_scaled1.values, (X_test_1.shape[0], 1600, 1))

# Define the model
model1_cnn = Sequential()

# Add a 1D Convolutional layer
model1_cnn.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(1600, 1)))
model1_cnn.add(MaxPooling1D(pool_size=2))
model1_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model1_cnn.add(MaxPooling1D(pool_size=2))
model1_cnn.add(Flatten())  # Flatten the output for the dense layers

# Add dense layers
model1_cnn.add(Dense(512, activation='relu'))
model1_cnn.add(Dense(256, activation='relu'))
model1_cnn.add(Dense(128, activation='relu'))

# Output layer with 5 classes (assuming 5 different activities)
model1_cnn.add(Dense(5, activation='softmax'))

# Compile the model with run_eagerly=True
model1_cnn.compile(optimizer=Adam(), loss=SparseCategoricalCrossentropy(), metrics=['accuracy'], run_eagerly=True)

# Print model summary
model1_cnn.summary()

# Train the model
history1_cnn = model1_cnn.fit(X_train_9_all_reshaped, y_train_9_all.values, epochs=100, batch_size=128, validation_data=(X_test_9_all_reshaped, y_test_9_all.values))
```

# 1.1 One volunteer

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rfc_1=RandomForestClassifier(max_depth=12)
classifier_rfc_1.fit(X_train_1,y_train_1)
y_pred_rfc_1=classifier_rfc_1.predict(X_train_1)
print("the accuracy of training data: ", accuracy_score(y_train_1,y_pred_rfc_1))
y_pred_rfc_test_1=classifier_rfc_1.predict(X_test_1)
print("the accuracy of testing data:  ", accuracy_score(y_test_1,y_pred_rfc_test_1))
RF_Accuracy.append(accuracy_score(y_test_1,y_pred_rfc_test_1))
```

```python
cm_rfc_1 = confusion_matrix(y_test_1, y_pred_rfc_test_1, normalize='true')
class_names = ["No_person","Down","Left","Right", "Up"]

# Plot confusion matrix as heatmap

sns.heatmap(cm_rfc_1, annot=True, cmap="Blues", fmt=".2%", vmin=0, vmax=1,xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix for Random Forest Classifier of one volunteer")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```

```python
report_rfc_1 = classification_report(y_test_1, y_pred_rfc_test_1)
print("Classification report for Random Forest of one volunteers:")
print(report_rfc_1)
```
