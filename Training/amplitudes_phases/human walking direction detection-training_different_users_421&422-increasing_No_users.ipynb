```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.1 One volunteer

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rfc_1=RandomForestClassifier(max_depth=12)
classifier_rfc_1.fit(X_train_1,y_train_1)
y_pred_rfc_1=classifier_rfc_1.predict(X_train_1)
print("the accuracy of training data: ", accuracy_score(y_train_1,y_pred_rfc_1))
y_pred_rfc_test_1=classifier_rfc_1.predict(X_test_1)
print("the accuracy of testing data:  ", accuracy_score(y_test_1,y_pred_rfc_test_1))
RF_Accuracy.append(accuracy_score(y_test_1,y_pred_rfc_test_1))
```

```python
cm_rfc_1 = confusion_matrix(y_test_1, y_pred_rfc_test_1, normalize='true')
class_names = ["No_person","Down","Left","Right", "Up"]

# Plot confusion matrix as heatmap

sns.heatmap(cm_rfc_1, annot=True, cmap="Blues", fmt=".2%", vmin=0, vmax=1,xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix for Random Forest Classifier of one volunteer")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```

```python
report_rfc_1 = classification_report(y_test_1, y_pred_rfc_test_1)
print("Classification report for Random Forest of one volunteers:")
print(report_rfc_1)
```
