# Train the dataset

```python
#Data encoding: to convert categorical features to numberical
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
dataset1['label']=label_encoder.fit_transform(dataset1['label'])
dataset1
```

```python
dataset2['label']=label_encoder.fit_transform(dataset2['label'])
dataset2
```

```python
dataset['label']=label_encoder.fit_transform(dataset['label'])
dataset
```

```python
dataset['label'].unique()
```

```python
y=dataset['label']
X=dataset.drop('label',axis=1)
```

```python
y_1=dataset1['label']
X_1=dataset1.drop('label',axis=1)
```

```python
y_2=dataset2['label']
X_2=dataset2.drop('label',axis=1)
```

```python
#Trainset Splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3,random_state=123)
X_train_1, X_test_1, y_train_1, y_test_1= train_test_split(X_1, y_1, test_size=0.3,random_state=123)
X_train_2, X_test_2, y_train_2, y_test_2= train_test_split(X_1, y_1, test_size=0.3,random_state=123)
```

# classification using ML models

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
RF_Accuracy=[]
SVM_Accuracy=[]
KNN_Accuracy=[]
```

# 1. Random Forest Classifier

# 1.1 Room Lab1

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rfc_1=RandomForestClassifier(max_depth=12)
classifier_rfc_1.fit(X_train_1,y_train_1)
y_pred_rfc_1=classifier_rfc_1.predict(X_train_1)
print("the accuracy of training data: ", accuracy_score(y_train_1,y_pred_rfc_1))
y_pred_rfc_test_1=classifier_rfc_1.predict(X_test_1)
print("the accuracy of testing data:  ", accuracy_score(y_test_1,y_pred_rfc_test_1))
RF_Accuracy.append(accuracy_score(y_test_1,y_pred_rfc_test_1))
```

```python
cm_rfc_1 = confusion_matrix(y_test_1, y_pred_rfc_test_1, normalize='true')
class_names = ["Empty","Down","Left","Right", "Up"]

# Plot confusion matrix as heatmap

sns.heatmap(cm_rfc_1, annot=True, cmap="Blues", fmt=".2%", vmin=0, vmax=1,xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix of the Random Forest Classifier in Room Lab1")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```

```python
report_rfc_1 = classification_report(y_test_1, y_pred_rfc_test_1)
print("The Classification report for Random Forest of Room Lab1:")
print(report_rfc_1)
```
